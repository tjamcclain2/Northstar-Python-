```python
# --- 0. INSTALL DEPENDENCIES ---
!pip install pycbc tqdm --quiet

# --- 1. IMPORTS ---
import requests
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import tensorflow as tf
from tensorflow.keras import layers, models
from pycbc.waveform import get_td_waveform

# --- 2. GPU CHECK ---
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    print(f"✅ Using GPU: {physical_devices}")
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
else:
    print("⚠️ No GPU detected! Go to Runtime > Change runtime type > GPU.")

# --- 3. LOAD GWTC CATALOG ---
def load_gwtc_events():
    url = "https://gwosc.org/eventapi/jsonfull/GWTC/"
    resp = requests.get(url)
    resp.raise_for_status()
    data = resp.json()
    events = []
    for key, info in data["events"].items():
        try:
            m1 = info["mass_1_source"]
            m2 = info["mass_2_source"]
            dist = info["luminosity_distance"]
            chi = info.get("chi_eff", 0.0) or 0.0
            if None not in (m1, m2, dist):
                events.append({
                    "name": key.split("-")[0],
                    "mass1": m1, "mass2": m2,
                    "distance": dist,
                    "chi_eff": chi
                })
        except:
            continue
    print(f"✅ Loaded {len(events)} events.")
    return events

# --- 4. TRAINING DATA GENERATION ---
def build_training_data(events, nsamp=1000, duration=2.0, fs=16384):
    N = int(duration * fs)
    delta_t = 1 / fs
    f_lower = 20.0

    X = np.zeros((nsamp, N, 1), dtype=np.float32)
    Y = np.zeros((nsamp, N, 1), dtype=np.float32)

    def add_realistic_noise(sig, t, sigma):
        # White Gaussian noise
        white = np.random.normal(0, sigma * np.random.uniform(0.5, 1.5), size=N)
        # Heavy-tailed noise (Student-t)
        heavy = np.random.standard_t(3, size=N) * sigma * np.random.uniform(0.3, 1.0)
        # Non-stationary envelope
        env = 1 + 0.5 * np.sin(2 * np.pi * 0.1 * t + np.random.rand()*10)
        heavy *= env
        # Random glitches
        glitches = np.zeros(N)
        for _ in range(np.random.poisson(3)):
            tg = np.random.uniform(0, t[-1])
            amp = sigma * np.random.uniform(0.5, 2.0)
            wid = np.random.uniform(0.001, 0.02)
            freq = np.random.uniform(50, 300)
            glitches += amp * np.exp(-0.5*((t - tg)/wid)**2) * np.sin(2*np.pi*freq*(t - tg))
        return sig + white + heavy + glitches

    print("⏳ Generating training data...")
    for i in tqdm(range(nsamp)):
        ev = np.random.choice(events)
        try:
            hp, _ = get_td_waveform(
                approximant="SEOBNRv4",
                mass1=ev["mass1"], mass2=ev["mass2"],
                spin1z=ev["chi_eff"], spin2z=0.0,
                distance=ev["distance"],
                delta_t=delta_t, f_lower=f_lower
            )
            clean = hp.data
            if len(clean) > N:
                clean = clean[:N]
            else:
                pad = N - len(clean)
                clean = np.pad(clean, (pad//2, pad - pad//2))
            t = np.arange(N) / fs
            sigma = np.std(clean) * np.random.uniform(0.8, 1.2)  # random SNR
            noisy = add_realistic_noise(clean, t, sigma)
            # Per-sample normalization
            max_val = np.max(np.abs(clean))
            if max_val == 0: max_val = 1
            X[i,:,0] = noisy / max_val
            Y[i,:,0] = clean / max_val
        except:
            continue
    return X, Y, t

# --- 5. SKIP-CONNECTION AUTOENCODER ---
def build_autoencoder(N):
    inp = layers.Input((N, 1))
    c1 = layers.Conv1D(16, 31, activation='relu', padding='same')(inp)
    p1 = layers.MaxPooling1D(2, padding='same')(c1)
    c2 = layers.Conv1D(8, 15, activation='relu', padding='same')(p1)
    p2 = layers.MaxPooling1D(2, padding='same')(c2)

    c3 = layers.Conv1D(8, 15, activation='relu', padding='same')(p2)
    u1 = layers.UpSampling1D(2)(c3)
    m1 = layers.Concatenate()([u1, c2])
    c4 = layers.Conv1D(16, 31, activation='relu', padding='same')(m1)
    u2 = layers.UpSampling1D(2)(c4)
    m2 = layers.Concatenate()([u2, c1])
    out = layers.Conv1D(1, 1, activation='linear', padding='same')(m2)

    model = models.Model(inp, out)
    model.compile(optimizer='adam', loss='mse')
    return model

# --- 6. RUN TRAINING ---
events = load_gwtc_events()
X, Y, t = build_training_data(events, nsamp=1200, duration=2.0, fs=16384)
N = X.shape[1]
split = int(0.8 * len(X))
X_tr, Y_tr = X[:split], Y[:split]
X_te, Y_te = X[split:], Y[split:]

model = build_autoencoder(N)
model.summary()
model.fit(X_tr, Y_tr, epochs=50, batch_size=8, validation_split=0.1, verbose=2)

# --- 7. MULTI-PLOT TEST RESULTS ---
n_plots = 3
idxs = np.random.choice(len(X_te), n_plots, replace=False)
plt.figure(figsize=(12, 6))
for k, i in enumerate(idxs, 1):
    noisy = X_te[i,:,0]
    den = model.predict(noisy[None,:,None])[0,:,0]
    clean = Y_te[i,:,0]
    plt.subplot(n_plots, 1, k)
    plt.plot(t, clean, label='Clean', lw=1)
    plt.plot(t, noisy, label='Noisy', alpha=0.5)
    plt.plot(t, den, label='Denoised', lw=1)
    plt.legend(loc='upper right')
    plt.ylabel("Strain")
plt.xlabel("Time [s]")
plt.suptitle("GWTC Events – Clean vs Noisy vs Denoised (16 kHz, Realistic Noise)")
plt.tight_layout()
plt.show()
```
